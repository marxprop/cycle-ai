```python
path = "cycle"
dls = ImageDataLoaders.from_folder(path, valid_pct = 0.1, item_tfms=Resize(460),batch_tfms=aug_transforms(size=224))
```

```python
dls.show_batch()
```

```python
dls.show_batch()
```

```python
learn = vision_learner(dls, resnet34, metrics=error_rate)
```

```python
learn.fine_tune(15)
```

```python
learn.export('model.pkl')
```
